```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
titanic=sns.load_dataset('titanic')
```

```python
titanic.head(30)
```

```python
df=titanic.copy()
```

```python
df.isna().sum()
```

```python
df['age'].fillna(df.age.mean(),inplace=True)
```

```python
df['deck'] = df['deck'].cat.add_categories('other')
df['deck'].fillna('other',inplace=True)
```

```python
df.isna().sum()
```

```python
sns.barplot(x='survived',y='who',hue='sex',data=df);
```

```python
sns.catplot(y='survived',x='sex',data=df)
```

```python
df.groupby(['sex','who','survived'])['age'].describe().T
```

```python
sns.distplot(df.fare)
```

```python
from pandas.api.types import CategoricalDtype
```

```python
classes=['Third','Second','First']
```

```python
df.rename(columns={'class':'class_types'},inplace=True)
```

```python
df.class_types=df.class_types.astype(CategoricalDtype(categories=classes,ordered=True))
```

```python
df.class_types.head()
```

```python
sns.FacetGrid(df,hue='class_types',height=10,xlim = (0, 1)).map(sns.kdeplot,'survived',shade=True).add_legend();
```

```python
sns.catplot(data=df, x='class_types',y='survived',hue='who',kind='point')
```

```python
fig = plt.figure(figsize=(10, 6))
sns.scatterplot(data=df,x='age',y='fare',hue='class_types',style='sex')
```

```python
fig = plt.figure(figsize=(60, 20))
sns.lineplot(x = "fare", y = "age", hue = "sex", data = df);
```
